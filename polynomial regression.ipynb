# Polynomial Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
data={
    'stuhr':[1,2,3,4,5],
    'marks':[10,25,55,90,130]
}
df=pd.DataFrame(data)
df
```

```python
x=df[['stuhr']]
y=df['marks']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4,random_state=42)
```

```python
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
model=LinearRegression()
model.fit(x_train_poly,y_train)
```

```python
y_train_pr=model.predict(x_train_poly)
y_test_pr=model.predict(x_test_poly)
```

```python
print("Training MSE:",mean_squared_error(y_train,y_train_pr))
print("Testing MSE:", mean_squared_error(y_test,y_test_pr))
```

```python
plt.scatter(x,y,color="blue",label="actual data")
xline=np.linspace(min(x["stuhr"]),max(x["stuhr"]),100).reshape(-1,1)
yline=model.predict(poly.transform(xline))
plt.plot(xline,yline,color='red',label="polynomial fit")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("polynomail regression")
plt.legend()
plt.show()
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```
