```python
for i in range (0,5,1):          #for loop
    print(i)
```

```python
n=int(input("enter a number:"))
for i in range(0,(n+1)):
    print(i)
    
```

```python
n=int(input("enter a number:"))# while comdition
i=0
while(i<=n):
    print(i)
    i+=1
```

```python
for i in range(5):#pass
    if i==3:
        pass
    print(i)
```

```python
for i in range(5):#continue
    if i==3:
        continue
    print(i)
```

```python
for i in range(5):#break
    if i==3:
        break
    print(i)
```
