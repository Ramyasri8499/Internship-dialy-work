# Train_test_split

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data=pd.DataFrame({
    "study_hours":[1,2,3,4,5,6,7,8],
    "marks":[35,40,50,55,65,70,80,85]
})
data
```

```python
x=data[["study_hours"]]  #input
y=data["marks"] #output
x
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=40)   
#here test_size is 20% sp it take 80% for train if we take train_test=0.80 and it take 20% to test automatically. 
# rando  state is used to suffle the data
```
